# Text Representation

**Text Representation** in Natural Language Processing (NLP) is the process of converting textual data into a numerical format that can be understood and processed by machine learning algorithms. It involves transforming raw text into a structured representation that captures the semantic and syntactic information contained within the text.

## Transforming Raw Text into Structured Representation

**Common text representation techniques:**

* **Bag of Words (BoW):** Represents a document as a numerical vector where each element corresponds to the frequency of a word in the document.
* **TF-IDF (Term Frequency-Inverse Document Frequency):** Assigns weights to words based on their frequency in a document and the corpus, giving more importance to informative words.
* **Word Embeddings:** Represents words as dense, continuous vectors in a high-dimensional space, capturing semantic relationships between words.
* **Character-level Embeddings:** Represents individual characters as numerical vectors, useful for handling rare words or languages with limited vocabulary.
* **Document Embeddings:** Represents entire documents as numerical vectors, capturing the overall semantic meaning of the document.

The choice of text representation technique depends on the specific NLP task and the characteristics of the data. For example, BoW is simple but may not capture word order or semantic relationships, while word embeddings can capture more complex linguistic information.

**Understanding the Goal:**

The primary objective is to convert raw text, which is a sequence of characters, into a numerical representation that machines can understand and process. This representation should capture the underlying semantic and syntactic information, such as the meaning of words, their relationships, and the structure of the text.

Let's consider the following sentence:

> "The quick brown fox jumps over the lazy dog."

### 1. **Tokenization:**

* Break the sentence into individual words or tokens:
  * `The`, `quick`, `brown`, `fox`, `jumps`, `over`, `the`, `lazy`, `dog`

### 2. **Text Cleaning:**

* Remove unwanted characters (e.g., punctuation, stop words):
  * `The`, `quick`, `brown`, `fox`, `jumps`, `over`, `lazy`, `dog`

### 3. **Text Normalization:**

* Convert words to a consistent form (e.g., lowercase):
  * `the`, `quick`, `brown`, `fox`, `jumps`, `over`, `the`, `lazy`, `dog`

### 4. **Text Representation:**

* **Bag of Words (BoW):**
  * Create a vocabulary: `the`, `quick`, `brown`, `fox`, `jumps`, `over`, `lazy`, `dog`
  * Represent the sentence as a numerical vector:
    * `[1, 1, 1, 1, 1, 1, 1, 1]` (each element represents the frequency of a word in the sentence)
* **TF-IDF:**
  * Calculate TF-IDF weights for each word based on its frequency in the sentence and the corpus.
  * Represent the sentence as a numerical vector with TF-IDF weights.
* **Word Embeddings:**
  * Use a pre-trained word embedding model (e.g., Word2Vec, GloVe) to represent each word as a dense vector.
  * The vector captures semantic relationships between words.
  * Represent the sentence as a sequence of these word vectors.

**Capturing Semantic and Syntactic Information:**

* **BoW:** Captures word frequency but ignores word order and semantic relationships.
* **TF-IDF:** Captures word importance based on frequency but still ignores word order.

* **N-grams:** Captures word order and phrases based on probability theory.

* **Word Embeddings:** Captures semantic relationships between words, allowing for understanding of meaning and context.

* **Document Embeddings:** Represent entire documents as vectors, capturing the overall semantic meaning.

* **Deep Learning Models:** Utilize deep learning models (e.g., recurrent neural networks, transformers) to capture complex linguistic patterns and dependencies.

By combining these techniques, you can create a structured representation of text that effectively captures the semantic and syntactic information, enabling machines to understand and process natural language.


## I. Tokenization

### Example 1: `string.split()`

**Purpose:**

* Splits a string into a list of substrings based on a specified separator.

**Syntax:**

```python
string.split(sep=None, maxsplit=-1)
```

**Parameters:**

* `sep`: Optional. The separator used to split the string. If not specified, whitespace characters (spaces, tabs, newlines) are used as separators.
* `maxsplit`: Optional. The maximum number of splits to perform. If not specified, all possible splits are made.

**Return Value:**

* A list of substrings resulting from the splitting operation.

**Examples:**

```python
text = "Hello, world!"

# Split using a comma as the separator:
words = text.split(",")
print(words)  # Output: ['Hello', ' world!']

# Split using whitespace as the separator:
words = text.split()
print(words)  # Output: ['Hello', 'world!']

# Split into at most two substrings:
words = text.split(maxsplit=1)
print(words)  # Output: ['Hello', ' world!']
```

**Key Points:**

* If `sep` is not specified, the string is split on whitespace characters.
* The `maxsplit` parameter controls the maximum number of splits.
* The returned list does not include the separator.
* If the separator is found at the beginning or end of the string, empty substrings are included in the result.

**Additional Notes:**

* You can use regular expressions for more complex splitting patterns using the `re.split()` function.
* If you need to split a string into a list of lines, you can use the `splitlines()` method.

By understanding the `str.split()` method, you can effectively manipulate and process strings in your Python programs.

### Example 2: `re.split()`

Using Regular Expressions `re.split()` function

**Purpose:**

* Splits a string into a list of substrings based on a regular expression pattern.

**Syntax:**

```python
re.split(pattern, string, maxsplit=0, flags=0)
```

**Parameters:**

* `pattern`: The regular expression pattern used for splitting.
* `string`: The input string to be split.
* `maxsplit`: Optional. The maximum number of splits to perform. If not specified, all possible splits are made.
* `flags`: Optional. Flags that modify the behavior of the regular expression.

**Return Value:**

* A list of substrings resulting from the splitting operation.

**Examples:**

```python
import re

text = "Hello, world! How are you?"

# Split using a comma as the separator:
words = re.split(r",", text)
print(words)  # Output: ['Hello', ' world! How are you?']

# Split using whitespace as the separator:
words = re.split(r"\s+", text)
print(words)  # Output: ['Hello', 'world!', 'How', 'are', 'you?']

# Split into at most two substrings:
words = re.split(r"\s+", text, maxsplit=1)
print(words)  # Output: ['Hello', 'world! How are you?']

# Split using a custom pattern:
words = re.split(r"[.,!?]", text)
print(words)  # Output: ['Hello', ' world', ' How', ' are', ' you', '']
```

**Key Points:**

* The `re.split()` function uses regular expressions to define the splitting pattern.
* The `maxsplit` parameter controls the maximum number of splits.
* The returned list does not include the matching pattern.
* Regular expressions provide a powerful and flexible way to specify splitting patterns.

**Additional Notes:**

* For more complex splitting patterns, you can use regular expression syntax to match specific characters, patterns, or groups.
* If you need to split a string into a list of lines, you can use the `re.split()` function with the pattern `r"\n"`.

By understanding the `re.split()` function, you can effectively manipulate and process strings in your Python programs using regular expressions.

## III. Text Cleaning

**Text Cleaning** is a crucial preprocessing step in Natural Language Processing (NLP) that involves removing noise, inconsistencies, and irrelevant information from text data. It helps to improve the quality and consistency of the data, making it more suitable for downstream tasks like sentiment analysis, machine translation, and text summarization.

**Common text cleaning techniques include:**

* **Removing punctuation:** Removing symbols like commas, periods, question marks, and exclamation points.
* **Handling contractions:** Expanding contractions like "can't" to "cannot" and "don't" to "do not."
* **Correcting spelling errors:** Identifying and correcting misspelled words.
* **Removing special characters:** Removing non-text characters like emojis, symbols, and control characters.
* **Handling HTML tags:** Removing HTML tags from text extracted from web pages.
* **Dealing with noise:** Removing noise like typos, OCR errors, or formatting inconsistencies.

**Example:**

Before cleaning: "I can't believe this is happening! It's so awesome!"

After cleaning: "I cannot believe this is happening it is so awesome"

By applying text cleaning techniques, you can improve the quality and consistency of your text data, making it more suitable for NLP tasks and achieving better results.


### Expanding Contractions


````python
import re

def expand_contractions(text):
  contractions = {
    "can't": "cannot",
    "it's": "it is",
  }

  # Regular expression to match contractions
  contraction_re = re.compile('(\w+[\'.\'|\"])(\w+)')

  def replace_contraction(match):
    expanded = contractions.get(match.group(0))
    return expanded

  return contraction_re.sub(replace_contraction, text)

# Example usage:
text = "I can't believe this is happening! It's so awesome!"

lowercase_text = text.lower()
print(f"Lowercase Text: {lowercase_text}\n")

expanded_text = expand_contractions(lowercase_text)
print(f"Expanded Text: {expanded_text}\n")  # Output: I cannot believe this is happening! It is so awesome!
````

## Text Preprocessing at once


````python
import re

def expand_contractions(text):
  contractions = {
    "can't": "cannot",
    "it's": "it is",
  }
  contraction_re = re.compile('(\w+[\'.\'|\"])(\w+)')
  def replace_contraction(match):
    expanded = contractions.get(match.group(0))
    return expanded

  return contraction_re.sub(replace_contraction, text)

def remove_punctuation(text):
  return re.sub(r'[^\w\s]', '', text)

def preprocess_text(text):
  lowercase_text = text.lower()
  expanded_text = expand_contractions(lowercase_text)
  cleaned_text = remove_punctuation(expanded_text)
  return cleaned_text.split()

# Example usage:
text = "I can't believe this is happening! It's so awesome!"

preprocessed_text = preprocess_text(text)
print(f"Preprocessed Text: {preprocessed_text}")
````

# Text Representation: BOW

**Bag of Words (BoW)**: Represents a document as a numerical vector where each element corresponds to the frequency of a word in the document.

> This representation assumes that the words in texts are independent of each other in accourance. That is, BOW essentially treats the document as a collection of words, ignoring the order in which they appear.

### The Independence Assumption
The core assumption behind BoW is that the occurrence of one word in a document is independent of the occurrence of another word. This means that the presence or absence of a word does not influence the probability of another word appearing.

**Example:**
Consider the two sentences:
* `"The quick brown fox jumps over the lazy dog."`
* `"The lazy dog jumps over the quick brown fox."`

In a BoW representation, these sentences would be identical because they contain the same words with the same frequencies. The order of the words is disregarded.

### Limitations of the Independence Assumption
While the BoW model is simple and often effective, it has limitations due to its independence assumption:

* **Word Order:** The order of words in a sentence can significantly affect its meaning. For example, "The dog bit the man" and "The man bit the dog" have different meanings, but a BoW representation would treat them as identical.
* **Semantic Relationships:** BoW does not capture semantic relationships between words. For instance, the words "cat" and "kitten" are semantically related, but a BoW model would treat them as distinct entities.
* **Contextual Information:** BoW ignores the context in which words appear. This can be problematic for tasks like sentiment analysis, where the meaning of a word can depend on its surrounding context.

### Addressing the Limitations
To address these limitations, more sophisticated techniques have been developed:

* **N-grams:** Consider sequences of words (n-grams) to capture some aspects of word order.
* **Word Embeddings:** Represent words as dense vectors in a high-dimensional space, capturing semantic relationships between words.

While BoW remains a fundamental technique in NLP, understanding its limitations and exploring more advanced methods is essential for achieving better results in many applications.


````python
from collections import Counter

def generate_bow_vector(text):
  preprocessed_text = preprocess_text(text)
  word_counts = Counter(preprocessed_text)
  vocabulary = sorted(list(set(preprocessed_text)))
  bow_vector = [word_counts[word] for word in vocabulary]
  return bow_vector, vocabulary

text = "Natural Language Processing (NLP) is a field of AI that focuses on the interaction between computers and human language."
bow_vector, vocabulary = generate_bow_vector(text)
print(f"Vocabulary: {vocabulary}")
print(f"BoW Vector: {bow_vector}")
````

# Zipf's Law

**Zipf's Law** is a linguistic law that states that the frequency of a word in a given text is inversely proportional to its rank in the frequency table. In other words, the most frequent word will appear approximately twice as often as the second most frequent word, and so on.

**Implications of Zipf's Law:**

* **Word Frequency Distribution:** Zipf's law helps explain the skewed distribution of word frequencies in natural languages, with a small number of words occurring very frequently and a large number of words occurring infrequently.
* **Language Modeling:** Zipf's law is often used in language modeling to estimate the probability of words occurring in a sequence.
* **Information Retrieval:** Understanding Zipf's law can help optimize information retrieval systems by focusing on indexing the most frequent terms.

To model Zipf's law, we can fit a power-law distribution to the frequency distribution of the words. Zipf's law can be expressed mathematically as:

$$
f(r) = \frac{C}{r^s}
$$

Where:
- $ f(r) $ is the frequency of the word with rank $ r $,
- $ C $ is a constant (related to the total frequency),
- $ r $ is the rank of the word,
- $ s $ is the exponent, typically close to 1.

### To make the fit
1. **Log-transform the rank and frequency data**: This converts the power law into a linear form:
   $$
   \log(f(r)) = \log(C) - s \log(r)
   $$
2. **Fit a line** to the log-transformed data using linear regression.
3. **Extract the slope** $ s $ and intercept $ \log(C) $ from the regression to give us the fitted equation.

### Code Explanation:
1. **Preprocessing**: Same as the previous step, the text is tokenized, cleaned, and lowercased.
2. **Word Frequencies**: We count the word frequencies and sort them by rank.
3. **Curve Fitting**:
   - We define the Zipf law function $ f(r) = \frac{C}{r^s} $ and use `curve_fit` from `scipy.optimize` to fit the data.
   - The `curve_fit` function returns the constants $ C $ and $ s $.
4. **Plot**: The log-log plot shows the data points and the fitted curve.



````python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from collections import Counter
import nltk
import re
import requests

# Ensure nltk data is available
nltk.download('punkt')

# Step 1: Download the dataset (Alice in Wonderland from Project Gutenberg)
url = 'https://www.gutenberg.org/files/11/11-0.txt'
response = requests.get(url)
text = response.text

# Step 2: Preprocess the text - remove non-alphabetic characters and tokenize
def preprocess(text):
    # Remove non-alphabetic characters and lowercase
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.lower()
    # Tokenize the text
    tokens = nltk.word_tokenize(text)
    return tokens

tokens = preprocess(text)

# Step 3: Calculate word frequencies
word_freq = Counter(tokens)

# Step 4: Sort by frequency (most common first)
sorted_word_freq = word_freq.most_common()

# Step 5: Rank words and get their frequencies
ranks = np.arange(1, len(sorted_word_freq) + 1)
frequencies = np.array([freq for _, freq in sorted_word_freq])

# Step 6: Define the Zipf function to fit: f(r) = C / r^s
def zipf_law(rank, C, s):
    return C / (rank ** s)

# Step 7: Use curve_fit to fit the Zipf law model to the data
popt, pcov = curve_fit(zipf_law, ranks, frequencies)

C, s = popt  # Extract the constants

# Step 8: Plot the data and the fitted curve
plt.figure(figsize=(10, 6))
plt.loglog(ranks, frequencies, marker=".", linestyle='none', label="Data")
plt.loglog(ranks, zipf_law(ranks, *popt), label=f"Fitted curve: C = {C:.2f}, s = {s:.2f}")
plt.title("Zipf's Law: Word Frequency Distribution with Fitted Curve for Alice in Wonderland")
plt.xlabel("Rank (log scale)")
plt.ylabel("Frequency (log scale)")
plt.legend()
plt.grid(True)
plt.show()

# Output the fitted equation
print(f"Fitted equation: f(r) = {C:.2f} / r^{s:.2f}")
````
